```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("props.txt", sep="\t")
df.plot(x="t", y="U", marker=".", ls="-")
df["U_ma"] = df["U"].rolling(5).mean()
df.dropna(axis=0, inplace=True)
df.to_csv("ma.txt", sep="\t", columns=["t", "U_ma"])

plt.plot(df["t"], df["U_ma"])
```
